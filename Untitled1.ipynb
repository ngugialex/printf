```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import plotly.express as px
```

```python
data = pd.read_csv('/content/Mobile phone price.csv')
```

```python
print(data.head())
```

```python
print(data.describe())
```

```python
data = pd.DataFrame({'numeric_column': [10, 15, 13, 17, 8, 12, 20, 7, 14, 16],
                     'categorical_column': ['A', 'B', 'A', 'B', 'A', 'C', 'C', 'B', 'C', 'A']})
```

```python
x = [1, 2, 3, 4, 5]
y = [10, 15, 13, 17, 8]
plt.plot(x, y)
plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')
plt.title('Line Plot')
plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y = [10, 15, 13, 17, 8]
plt.scatter(x, y)
plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')
plt.title('Scatter Plot')
plt.show()
```

```python
plt.figure(figsize=(8, 4))
sns.histplot(data['numeric_column'], kde=True)
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram')
plt.show()
```

```python
# Box Plot
plt.figure(figsize=(8, 6))
sns.boxplot(x=data['categorical_column'], y=data['numeric_column'])
plt.xlabel('Category')
plt.ylabel('Value')
plt.title('Box Plot by Category')
plt.xticks(rotation=45)
plt.show()
```

```python
data = pd.DataFrame({'x_column': [1, 2, 3, 4, 5],
                     'y_column': [10, 15, 13, 17, 8],
                     'category_column': ['A', 'B', 'A', 'B', 'A']})
```

```python
fig = px.scatter(data, x='x_column', y='y_column', color='category_column', title='Scatter Plot')
fig.show()
```

```python
data = pd.DataFrame({'x_column': [1, 2, 3, 4, 5],
                     'y_column': [10, 15, 13, 17, 8],
                     'category_column': ['A', 'B', 'A', 'B', 'A']})
```

```python
fig = px.bar(data, x='category_column', y='y_column', title='Bar Chart')
fig.show()
```
